```python
from sklearn.metrics import accuracy_score
```

```python
#Load the data set to a pandas dataframe
credit_card=pd.read_csv('/content/drive/MyDrive/creditcard.csv')
```

The dataset is highly unbalance

0----> liget data set
1----> fraudlent data set

```python
#now for analysis we will seprate the two data set and store the data in two diffrent variable
liget=credit_card[credit_card.Class==0]
fraudlent =credit_card[credit_card.Class==1]
```

```python
#now we will see the shape of the data
print(liget.shape)
print(fraudlent.shape)
```

```python
#Stastical measure of the given data
liget.Amount.describe()
```

```python
fraudlent.Amount.describe()
```

```python
#now we need to compare the value for both the transation
credit_card.groupby('Class').mean()
```

Now We will use here under sampling
Built a sample dataset containing similar distribution of normal transation and fraudlent transion
now the no. of fraudlent transation is 492 so from normal
transation we wilol take the 492 random sample

```python
legit_sample=liget.sample(n=492)
```

```python
#Now we will concatinate the data frame
new_dataset=pd.concat([legit_sample,fraudlent], axis=0)
```

```python
new_dataset
```

```python
new_dataset['Class'].value_counts()
```

```python
#now we will find the mean value of the new data set
new_dataset.groupby(['Class']).mean()
```

```python
#Splitting the data into the features and the target
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Split the data into the training and testing dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training
we are going to train the model using the logistic regression

```python
model=LogisticRegression()
```

```python
#training the logfistic regression model with the trainig data
model.fit(x_train,y_train)
```

Model Evaluation Accuracy score

```python
#accuracy on the training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy score is :',training_data_accuracy)
```

```python
#accuracy on the testing data
x_test_prediction=model.predict(x_test)
training_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score is :',training_data_accuracy)
```
